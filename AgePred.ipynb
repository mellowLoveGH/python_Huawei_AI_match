```python
user_basic_info=pd.read_csv("user_basic_info.csv",names=["uId", "gender", "city", "prodName","ramCapacity","ramLeftRation","romCapacity","romLeftRation","color","fontSize","ct","carrier","os"])
user_basic_info.tail(10)
```

```python
user_behavior_info=pd.read_csv("user_behavior_info.csv",names=["uId", "bootTimes", "AFuncTimes", "BFuncTimes","CFuncTimes","DFuncTimes","EFuncTimes","FFuncSum","GFuncSum"])
user_behavior_info.tail(10)
```

```python
to_predict=pd.merge(left=merged_inner,right=user_behavior_info, left_on='uId', right_on='uId')
to_predict.tail()
```

```python
to_predict=to_predict.dropna()
to_predict.reset_index()
to_predict.shape
```

# Feature engineering

```python
X_labels=["gender", "city","prodName","ramCapacity","ramLeftRation","romCapacity","romLeftRation","color","fontSize","ct","carrier","os","bootTimes", "AFuncTimes", "BFuncTimes","CFuncTimes","DFuncTimes","EFuncTimes","FFuncSum","GFuncSum"]
y_label='age_group'

from sklearn.preprocessing import LabelEncoder

lb_make = LabelEncoder()
to_predict["carrier"] = lb_make.fit_transform(to_predict["carrier"])
to_predict["color"] = lb_make.fit_transform(to_predict["color"])
to_predict["ct"] = lb_make.fit_transform(to_predict["ct"])
to_predict["prodName"] = lb_make.fit_transform(to_predict["prodName"])
to_predict["city"] = lb_make.fit_transform(to_predict["city"])
to_predict.head(11)
```

```python
to_predict.shape
```

```python
to_predict['age_group'].value_counts()
```

```python
y_data=to_predict[y_label]
X_data=to_predict[X_labels]
# del to_predict['age_group']
# to_predict.tail()
print(X_data.shape)
y_data.shape
```

```python
from sklearn import preprocessing
X = X_data.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
X_scaled = min_max_scaler.fit_transform(X)
X_scaled
# df = pandas.DataFrame(x_scaled)
```

```python
y=y_data.values
y
```

# Model Training

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=11)

# for experimental purpose, we only use 10% of data to train within a perceivable period of time
# X_train, X_test, y_train, y_test = train_test_split(X_test, y_test, test_size=0.2, random_state=11)
# X_train, X_test, y_train, y_test = train_test_split(X_test, y_test, test_size=0.2, random_state=11)

print(X_train.shape)
print(y_train.shape)
print("------------")
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.naive_bayes import GaussianNB
ml_m = GaussianNB()
%time ml_m.fit(X_train, y_train)
print('Accuracy of GaussianNB classifier on Test Set: {:.4f}'
     .format(ml_m.score(X_test, y_test)))

predicted_test = ml_m.predict(X_test)
from sklearn.metrics import classification_report
print(classification_report(y_test, predicted_test))
```

```python
from sklearn.svm import LinearSVC
lsvc_pri_model = LinearSVC(C=0.1)

%time lsvc_pri_model.fit(X_train, y_train)
print('Accuracy of LinearSVC classifier on Test Set: {:.4f}'
     .format(lsvc_pri_model.score(X_test, y_test)))

predicted_test = lsvc_pri_model.predict(X_test)
from sklearn.metrics import classification_report
print(classification_report(y_test, predicted_test))
```

# Prediction for test set and submission

```python
merged_inner_test = pd.merge(left=age_test,right=user_basic_info, left_on='uId', right_on='uId')
to_predict_ontest=pd.merge(left=merged_inner_test,right=user_behavior_info, left_on='uId', right_on='uId')
# to_predict_ontest.tail()
print(to_predict_ontest.shape)
to_predict_ontest=to_predict_ontest.dropna()
print(to_predict_ontest.shape)

from sklearn.preprocessing import LabelEncoder

lb_make = LabelEncoder()
to_predict_ontest["carrier"] = lb_make.fit_transform(to_predict_ontest["carrier"])
to_predict_ontest["color"] = lb_make.fit_transform(to_predict_ontest["color"])
to_predict_ontest["ct"] = lb_make.fit_transform(to_predict_ontest["ct"])
to_predict_ontest["prodName"] = lb_make.fit_transform(to_predict_ontest["prodName"])
to_predict_ontest["city"] = lb_make.fit_transform(to_predict_ontest["city"])
to_predict_ontest.head(11)
```

```python
test_labels=["gender", "city","prodName","ramCapacity","ramLeftRation","romCapacity","romLeftRation","color","fontSize","ct","carrier","os","bootTimes", "AFuncTimes", "BFuncTimes","CFuncTimes","DFuncTimes","EFuncTimes","FFuncSum","GFuncSum"]

test_data=to_predict_ontest[test_labels]

from sklearn import preprocessing
test = test_data.values #returns a numpy array

min_max_scaler = preprocessing.MinMaxScaler()
test_scaled = min_max_scaler.fit_transform(test)
test_scaled

# test_scaled=test
```

```python
predicted = lsvc_pri_model.predict(test_scaled)
predicted
```

```python
sub_df=pd.DataFrame(to_predict_ontest['uId']).reset_index(drop=True)
sub_df['label']=predicted
```
